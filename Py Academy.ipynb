```python
import pandas as pd
import numpy as np
```

```python
df_schools = pd.read_csv("schools_complete.csv")
df_students = pd.read_csv("students_complete.csv")
```

```python

#   * Total Schools
#   * Total Students
#   * Total Budget

print(len(df_schools['school_name'].unique()))
print(len(df_students['Student ID'].unique()))
print(df_students['Student ID'].count())
print(df_schools['budget'].sum(axis=0))
print(df_schools["type"].value_counts())
```

```python
# Place all of the data found into a summary DataFrame

total_schools = (len(df_schools['school_name'].unique()))
total_students = (len(df_students['Student ID'].unique()))
total_budget = (df_schools['budget'].sum(axis=0))
school_type = (df_schools["type"].value_counts())
```

```python
df_students.head(5)
df_schools.head()
```

```python
df_reading_scores = df_students["reading_score"].mean()
df_math_scores = df_students["math_score"].mean()


print(df_reading_scores)

print(df_math_scores)
```

```python
#   % Passing Math
#  % Passing Reading

df_math_scores_passing = [(df_students['math_score']>= 70)] 
df_reading_scores_passing = [(df_students['reading_score']>= 70)] 

df_passingscores_math = (np.mean(df_math_scores_passing))
df_passingscores_reading= (np.mean(df_reading_scores_passing))

df_average_passing_scores = np.mean([df_passingscores_math, df_passingscores_reading])*100
```

# District Summary

```python
summary_table_district = pd.DataFrame({"Total School count": [total_schools],
                              "Total Student count": [total_students],
                              "Total Budget": [total_budget],
                              "Average Reading score":[df_reading_scores],
                              "Average Math score":[df_math_scores],
                            "Average Passing Math score":[df_passingscores_math],
                            "Average Passing Reading score":[df_passingscores_reading],
                            "Average Passing scores":[df_average_passing_scores]})


summary_table_district
```

# Step 2

```python
# Merge two dataframes using an inner join
merge_table = pd.merge(df_schools, df_students ,on="school_name")

merge_table.head()
```

```python
#Create an overview table that summarizes key metrics about each school, including:

# School Name
# School Type
# Total Students
# Total School Budget
# Per Student Budget
# Average Math Score
# Average Reading Score
```

```python

# School Type



school_type = merge_table.groupby("school_name")["type"].unique()
#school_type = merge_table.groupby("school_name")["type"].count().reset_index()


               
print(school_type) 

#type_of_school.head()
```

```python



#Calculate the totals using school name as index



students_per_school =  merge_table.groupby("school_name").count()["School ID"]
Budget_per_school = df_schools.groupby("school_name").sum()["budget"]
Per_student_budget = Budget_per_school/students_per_school




students_per_school_df = pd.DataFrame({"Students Per School": students_per_school})




Budget_per_school_df = pd.DataFrame({"Budget_per_school": Budget_per_school})

Per_student_budget_df = pd.DataFrame({"Budget_per_student": Per_student_budget})


total_math_df = merge_table.groupby("school_name").sum()["math_score"]
average_math_df = total_math_df/students_per_school

average_math_df


total_reading_df = merge_table.groupby("school_name").sum()["reading_score"]
average_reading_df = total_reading_df/students_per_school

average_reading_df






# passing_math 

passing_math = merge_table.loc[merge_table["math_score"] >=70, ["school_name","math_score"]].groupby("school_name").count()


passing_scores_math = (passing_math["math_score"]/students_per_school)*100


# passing_reading
  
passing_reading = merge_table.loc[merge_table["reading_score"] >=70, ["school_name","reading_score"]].groupby("school_name").count()


passing_scores_reading = (passing_reading["reading_score"]/students_per_school)*100



# print(passing_scores_math)

# print(passing_scores_reading)
```

```python



Summary_Table_Schools = pd.DataFrame({"Students Per School": students_per_school, "Type": school_type,"School Budget":Budget_per_school, "Budget_per_student": Per_student_budget,
                                      "Average Math Score":average_math_df,"Average Reading Score": average_reading_df,"%Passing Math": passing_scores_math,
                                    "%Passing Reading": passing_scores_reading})

Summary_Table_Schools.head()


```

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

Summary_Table_Schools["budget_group"] = pd.cut(Summary_Table_Schools["Budget_per_student"],
                                              bins=spending_bins,
                                              labels=group_names)








   

Summary_Table_Schools_budget = Summary_Table_Schools.groupby("budget_group").mean()

Summary_Table_Schools_budget = Summary_Table_Schools_budget.drop(columns=['Students Per School','School Budget','Budget_per_student'])

Summary_Table_Schools_budget.head()  

```

#  Scores by School Size





```python
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

Summary_Table_Schools["size_group"] = pd.cut(Summary_Table_Schools["Students Per School"],
                                              bins= size_bins,
                                              labels=group_names)



Summary_Table_Schools_size = Summary_Table_Schools.groupby("size_group").mean()
Summary_Table_Schools_size = Summary_Table_Schools_size.drop(columns=['Students Per School','School Budget','Budget_per_student'])

Summary_Table_Schools_size.head()
```
